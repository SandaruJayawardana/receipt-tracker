```python
class Company:
    def __init__(self, name="", address="", abn="", tele=""):
        self.name = name
        self.address = address
        self.abn = abn
        self.tele = tele

class Customer:
    def __init__(self, name="", address="", tele=""):
        self.name = name
        self.address = address
        self.tele = tele

class Item:
    def __init__(self, name="", quantity=0, unit_price=0.0, total_price=0.0, other=""):
        self.name = name
        self.quantity = quantity
        self.unit_price = unit_price
        self.total_price = total_price
        self.other = other

class ReceiptBody:
    def __init__(self):
        self.products = []
        self.gst = 0.0
        self.total_with_tax = 0.0
        self.total = 0.0

    def add_item(self, item):
        self.products.append(item)

class ReceiptFoot:
    def __init__(self, card_number=""):
        self.card_number = card_number

class ReceiptHeading:
    def __init__(self, company=None, customer=None):
        if company is None:
            company = Company()
        if customer is None:
            customer = Customer()
        self.company = company
        self.customer = customer

class Receipt:
    def __init__(self, heading=None, body=None, foot=None):
        if heading is None:
            heading = ReceiptHeading()
        if body is None:
            body = ReceiptBody()
        if foot is None:
            foot = ReceiptFoot()
        self.heading = heading
        self.body = body
        self.foot = foot

    def display(self):
        print("Company:")
        print(f"  Name: {self.heading.company.name}")
        print(f"  Address: {self.heading.company.address}")
        print(f"  ABN: {self.heading.company.abn}")
        print(f"  Telephone: {self.heading.company.tele}")
        print("Customer:")
        print(f"  Name: {self.heading.customer.name}")
        print(f"  Address: {self.heading.customer.address}")
        print(f"  Telephone: {self.heading.customer.tele}")
        print("Products:")
        for item in self.body.products:
            print(f"  Name: {item.name}")
            print(f"  Quantity: {item.quantity}")
            print(f"  Unit Price: {item.unit_price}")
            print(f"  Total Price: {item.total_price}")
            print(f"  Other: {item.other}")
        print(f"GST: {self.body.gst}")
        print(f"Total with Tax: {self.body.total_with_tax}")
        print(f"Total: {self.body.total}")
        print(f"Card Number: {self.foot.card_number}")


```
